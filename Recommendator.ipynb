```python
import math
import numpy as np
import pandas as pd 
import scipy

song_matrix = pd.read_csv('songs.csv', sep=',')
song_matrix
```

```python
def get_centroid(songs):

    centroid = []
    i = 0
    for song in songs:
        if(i != 0):
            centroid.append(scipy.mean(song))
        i = i + 1
    return centroid
```

```python
my_songs = pd.read_csv('my_songs.csv', sep=',', error_bad_lines=False, encoding = 'latin-1')
my_songs
```

```python
import re

trans_songs = my_songs.T.to_numpy()

centroid = get_centroid(trans_songs)
songs = song_matrix.to_numpy()

distance_temp = []
for song in songs:
    i = 0
    distance = 0
    for x in song:
        if(i == 0):
            file_name = x
        else:
            distance = distance + (centroid[i - 1] - x) ** 2
        i = i + 1
    distance_temp.append((file_name, 1 - math.sqrt(distance)))

temp = sorted(distance_temp, key=lambda song: song[1], reverse=True)

for x in temp:
    print(re.sub(r'[mp3\d\.+]', '', x[0]).strip(), ": ", x[1])
```
